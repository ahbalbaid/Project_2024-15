```python
from matplotlib.ticker import ScalarFormatter
def plot_pandas(df, title, xlabel, ylabel, saved="output"):
    plot = df.plot(kind="line", 
        marker='o', 
        title=title, 
        xlabel=xlabel,
        ylabel=ylabel, 
        figsize=((12, 6)),
#         logy=True
    )        
#     plot.set_xscale('log', base=2)
#     plot.set_yscale('log', base=2)

#     plot.xaxis.set_major_formatter(ScalarFormatter())
#     plot.yaxis.set_major_formatter(ScalarFormatter())

#     plot.xaxis.set_minor_formatter(ScalarFormatter())
#     plot.yaxis.set_minor_formatter(ScalarFormatter())
#     plt.savefig(saved + "_log")
    plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
    plt.savefig(saved)
    plt.show()
    return plot
```

```python
specific_array_size = 268435456
```

```python
for node in nodes:
    plot_pandas(
            df=combined_df.loc[node, "Avg time/rank"].xs(specific_array_size, level="array_size").unstack(level="type"),
            title=f"{node.frame['name']} - Avg Time per Rank (Array Size: {specific_array_size})",
            xlabel="Processors",
            ylabel="Time (seconds)",
            saved= "./test"
        )
```
